```python
#定义LeNet网络结构
```

```python
class LeNet(nn.Module):#继承了nn.Module中的所有功能
    def __init__(self):
        super(LeNet, self).__init__()#启动父类：nn.Module的初始化程序
        # 使用 nn.Sequential 将图中的各层按顺序叠加
        self.net = nn.Sequential(
            # 1. 第一层卷积：输入通道1(灰度图), 输出通道6, 卷积核5x5, 填充2
            # 尺寸计算：(28 + 2*2 - 5) + 1 = 28。输出特征图：6 x 28 x 28
            nn.Conv2d(1, 6, kernel_size=5, padding=2),
            nn.Sigmoid(), # 经典 LeNet 使用 Sigmoid 激活函数

            # 2. 第一层池化：2x2 平均池化，步幅 2
            # 尺寸计算：28 / 2 = 14。输出特征图：6 x 14 x 14
            nn.AvgPool2d(kernel_size=2, stride=2),

            # 3. 第二层卷积：输入通道6, 输出通道16, 卷积核5x5 (无填充)
            # 尺寸计算：(14 - 5) + 1 = 10。输出特征图：16 x 10 x 10
            nn.Conv2d(6, 16, kernel_size=5),
            nn.Sigmoid(),

            # 4. 第二层池化：2x2 平均池化，步幅 2
            # 尺寸计算：10 / 2 = 5。输出特征图：16 x 5 x 5
            nn.AvgPool2d(kernel_size=2, stride=2),

            # 5. 展平层：将 3D 的特征图拉直成 1D 向量，为全连接层做准备
            # 计算：16通道 * 5高 * 5宽 = 400 个元素
            nn.Flatten(),

            # 6. 第一层全连接：输入400个神经元，输出120个神经元
            nn.Linear(400, 120),
            nn.Sigmoid(),

            # 7. 第二层全连接：输入120个神经元，输出84个神经元
            nn.Linear(120, 84),
            nn.Sigmoid(),

            # 8. 第三层全连接(输出层)：输入84个神经元，输出10个神经元(对应0-9数字)
            nn.Linear(84, 10)
        )

    def forward(self, x):
        for layer in self.net:
            x = layer(x)
            # 这一行是调试神器，让你看清数据在哪一步变小了
            print(f"层: {layer.__class__.__name__}, 形状: {x.shape}") 
        return x
```

```python
model = LeNet()
# 模拟一张 28x28 的输入图片 (BatchSize=1, Channel=1, H=28, W=28)
X = torch.randn(1, 1, 28, 28)
output = model(X)
print(f"最终输出形状: {output.shape}") # 预期应该是 [1, 10]
```
