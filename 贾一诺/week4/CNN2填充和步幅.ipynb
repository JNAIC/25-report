```python
import torch
from torch import nn
```

```python
#增加填充
```

```python
def comp_conv2d(conv2d,X):
    X=X.reshape((1,1)+X.shape)
    Y=conv2d(X)
    return Y.reshape(Y.shape[2:])

conv2d=nn.Conv2d(1,1,kernel_size=3,padding=1)#保持输入输出大小不变
X=torch.rand(size=(8,8))
comp_conv2d(conv2d,X).shape
```

```python
conv2d=nn.Conv2d(1,1,kernel_size=(5,3),padding=(2,1))#p=(k-1)/2
comp_conv2d(conv2d,X).shape
```

```python
#增加步幅
```

```python
conv2d=nn.Conv2d(1,1,kernel_size=3,padding=1,stride=2)#原始大小/步长
comp_conv2d(conv2d,X).shape
```

```python
conv2d=nn.Conv2d(1,1,kernel_size=(5,3),padding=(0,1),stride=(2,4))#p=(k-1)/2
comp_conv2d(conv2d,X).shape
```
