# 2.Jupyter

Jupyter的强大功能在于以下几点：

- 代码分块运行,每个文本框为一块,方便逐个运行,就像下面这样

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

train_df = pd.read_csv(r'data2\cleaned_encoded_train.csv')
test_df = pd.read_csv(r'data2\cleaned_encoded_test.csv')
origin_id = pd.read_csv(r'data2\test.csv')['client_id']

x_train = train_df.drop(columns=['y'])
y_train = train_df['y']
# 这里的代码单独运行用于读取和数据处理
```

```python
rf_model = RandomForestClassifier(n_estimators=300, max_depth=15, random_state=42, n_jobs=-1 )
rf_model.fit(x_train,y_train)

x_test = test_df
y_test = rf_model.predict(x_test)

label_map = {0: '<=50K', 1: '>50K'}
test_predictions_label = [label_map[p] for p in y_test]
# 这里的代码用于输出训练模型
```

```python
submission = pd.DataFrame({
    'client_id': origin_id,
    'y': test_predictions_label
})

submission_file = 'submission.csv'
submission.to_csv(submission_file,index = False)
# 用于输出
```

按住shift+enter即可运行
  

- 可以嵌入markdown与latex，使代码可读性与模块性大大增强
